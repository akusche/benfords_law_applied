Some basics intis. Sets url to be crawled and number of pages that will be crawled. In addition file *cars.p* is loaded if already exists. This files includes cars from last run.

```python
url = 'https://www.carforyou.ch/de/auto/search?page='
n = 5000 #defines number of cralwed pages
try:
    cars = pickle.load(open( "data/cars.p", "rb")) #reads history
except:
    cars = dict()
    print("No history available. You are crawling for the very fist time.")
```

```python
def extract_car(html):
    #input html code
    #output car list of cars with different paremeters (id, zip_code, price ...)
    
    #only the passages with cars are extracted from the entire HTML code
    soup = BeautifulSoup(page.content, 'html.parser')
    cars_html = soup.find_all('article', class_='card-link')

    #each car's attributs are extracted
    for c in cars_html:
        c_id = c.attrs['id']
        c_href = c.find('a').attrs['href']
        c_price = c.select('a p')[0].text
        c_header = c.select('h1')[0].text
        c_year_km = c.select('p')[-1].text
        c_seller = c.select('footer span')[0].text
        c_place = c.select('footer span')[-1].text
        c_insert_date = datetime.now()
       
        #clean data
        c_price = re.sub("[^0-9]", "", c_price)
        
        #,erge the individual attributes into a list
        c_list = (c_href, c_price, c_header, c_year_km, c_seller, c_place, c_insert_date)
        
        #add to the dictionary if it doesn't already exist
        cars[c_id] = c_list
```

```python
for i in range(1,n+1):
    #With every call, the page counter is increased so new cars are called up
    url_crawl = url + str(i)
    print(url_crawl)
    page = requests.get(url_crawl)
    
    #retrieved file is archived for optional debugging
    file = open("tmp/index" + str(i) + ".html", "wb") 
    file.write(page.content) 
    file.close() 
    
    #cars are added to the car list
    extract_car(page.content)    
```

```python
#saves crawled cars
pickle.dump( cars, open( "data/cars.p", "wb" ) )
```

```python
#save carwled cars to excel
df_cars = pd.DataFrame.from_dict(cars, 
                                orient='index', 
                                 columns=["href", "price", "header", "year_km", "seller", "place", "insert_date"])
df_cars.to_excel("output.xlsx") 
```
